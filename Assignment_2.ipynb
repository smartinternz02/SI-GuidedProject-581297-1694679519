```python
#Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the car crashes dataset
df = sns.load_dataset("car_crashes")
df.head()
```

**1. Scatter Plot**

```python
sns.scatterplot(x='total', y='alcohol', data=df)
plt.title('Scatter Plot between Total Crashes and Alcohol')
plt.xlabel('Total Crashes')
plt.ylabel('Alcohol-Related Crashes')
plt.show()
```

**Inference: There is a positive correlation between the total number of crashes and the number of alcohol-related crashes.**

**2. Histogram**

```python
plt.figure(figsize=(9,5))
sns.histplot(df['total'], bins=30)
plt.title('Histogram of Total Crashes')
plt.xlabel('Total Crashes')
plt.ylabel('Frequency')
plt.show()
```

**Inference: The majority of states have a total number of crashes below 20.**

**3. Box Plot**

```python
plt.figure(figsize=(8,5))
sns.boxplot(df.iloc[:,0:5])
plt.show()
```

**Inference: The boxplot of the dataset shows that there are few outliers for number of crashes due to alcohol and not_ditracted crashes.**

**4. Bar Plot**

```python
plt.figure(figsize=(9,5))
sns.barplot(x='abbrev', y='speeding', data=df)
plt.title('Average Speeding Crashes by State')
plt.xlabel('State Abbreviation')
plt.ylabel('Average Speeding Crashes')
plt.xticks(rotation=90)
plt.show()
```

**Inference: Some states have a significantly higher average number of speeding-related crashes.**

**5. Pair Plot**

```python
plt.figure(figsize=(9,5))
sns.pairplot(df[['total', 'speeding', 'alcohol']])
plt.suptitle('Pair Plot of Key Variables')
plt.show()
```

**Inference: It helps us visualize the relationships between key variables in the dataset.**

**6. Heatmap**

```python
plt.figure(figsize=(8,6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Inference: It helps us identify the strength and direction of relationships between variables.**

**7. Line Plot**

```python
plt.figure(figsize=(16,5))
sns.lineplot(x='ins_premium', y ='ins_losses', data=df, marker='.', markerfacecolor='limegreen', markersize=20)
plt.xlabel('ins_premium')
plt.ylabel('ins_losses')
plt.title('Premium - loss comparison')
plt.show()
```

**Inference: compares the insurance premium and insurance loss over the car crashes. The maximum loss occurred for a insurance premium of 1300 and minimum loss is occurred for an insurance premium of 650**

**8. Joint Plot**

```python
plt.figure(figsize=(8,6))
sns.jointplot(x='alcohol', y='speeding', data=df, kind='scatter')
plt.suptitle('Joint Plot of Alcohol vs. Speeding')
plt.show()
```

**Inference: It helps us see the relationship between alcohol-related crashes and speeding.**

**9. Violin Plot**

```python
plt.figure(figsize=(9,5))
sns.violinplot(x='abbrev', y='alcohol', data=df)
plt.title('Violin Plot of Alcohol-Related Crashes by State')
plt.xlabel('State Abbreviation')
plt.ylabel('Alcohol-Related Crashes')
plt.xticks(rotation=90)
plt.show()
```

**Inference: It provides a more detailed distribution of alcohol-related crashes by state.**

**10. Swarm Plot**

```python
plt.figure(figsize=(10,6))
sns.swarmplot(df.iloc[:,5:7])
plt.title("Distribution of Insurance premiums and insurance losses")
plt.ylabel("Amounts")
plt.show()
```

**Inference: It shows the distribution of ins_premium and ins_losses of the data. It shows that the losses are very small compared to the premium**
